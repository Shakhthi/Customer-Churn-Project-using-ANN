```python
import tensorflow as tf
from tensorflow.keras.models import load_model

import pandas as pd
import pickle
```

```python
X_test = pd.read_csv("data/X_test.csv")
y_test = pd.read_csv("data/y_test.csv")
```

```python
with open("artifacts/onehotencoder_geography.pkl", 'rb') as file:
    onehot_geo = pickle.load(file)

with open("artifacts/labelencoder_gender.pkl", "rb") as file:
    lableencoder_gender = pickle.load(file)

with open("artifacts/standardscaler.pkl", "rb") as file:
    stdscaler = pickle.load(file)

model = load_model("artifacts/ann_churn_model.keras")
```

```python
# Example input data
input_data = {
    'CreditScore': 600,
    'Geography': 'France',
    'Gender': 'Male',
    'Age': 40,
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 50000,
    'Geography_France': 1,
    'Geography_Germany': 0,
    'Geography_Spain': 0
}

input_df = pd.DataFrame([input_data])
```

```python
input_df["Gender"] = lableencoder_gender.transform(input_df["Gender"].values.reshape(-1, 1))
```

```python
input_df["Geography"] = onehot_geo.transform(input_df["Geography"].values.reshape(-1, 1))
```

```python
input_df_scaled = stdscaler.transform(input_df.drop("Geography", axis=1))
```

```python
model_prediction = model.predict(input_df_scaled)
print("Predicted Churn Probability:", model_prediction[0][0]*100, "%")
```
